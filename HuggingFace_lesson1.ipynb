```python
from datasets import load_dataset
from transformers import AutoModelForSeq2SeqLM
from transformers import AutoTokenizer
from transformers import GenerationConfig
```

```python
huggingface_dataset_name="knkarthick/dialogsum"
datset=load_dataset(huggingface_dataset_name)
```

```python
example_indeces=[40,200]
dash_line='-'.join(''for x in range(100))
for i, index in enumerate(example_indeces):
  print(dash_line)
  print("Example")
  print(dash_line)
  print("Input dialoge")
  print(datset['test'][index]['dialogue'])
  print(dash_line)
  print("Baseline Human Summary")
  print(datset['test'][index]['summary'])
  print(dash_line)
  print()
```

```python
model_name='google/flan-t5-base'
model=AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
tokenizer=AutoTokenizer.from_pretrained(model_name,use_fast=True)
```

```python
sentence="How you doing?"
sentence_encoded=tokenizer(sentence,return_tensors='pt')
sentence_decoded=tokenizer.decode(sentence_encoded['input_ids'][0],skip_special_tokens=True)
print("Encoded sentence\n")
print(sentence_encoded['input_ids'][0])
print("\nDecoded Sentence")
print(sentence_decoded)
```

```python
for i, index in enumerate(example_indeces):
  dialogue=datset['test'][index]['dialogue']
  summary=datset['test'][index]['summary']
  inputs=tokenizer(dialogue,return_tensors='pt')
  outputs=tokenizer.decode(model.generate(inputs['input_ids'],max_new_tokens=50,)[0],skip_special_Tokens=True)
  print(dash_line)
  print("Example",i+1)
  print(dash_line)
  print(f"Input prompt:\n{dialogue}")
  #print(datset['test'][index]['dialogue'])
  print(dash_line)
  print(f"Baseline Human Summary:\n{summary}")
  #print(datset['test'][index]['summary'])
  print(dash_line)
  print(f"Model generation-without promt engineering:\n{outputs}")
```

```python
def make_prompt(example_indeces_full,example_index_to_summarize):
  prompt=""
  for index in example_indeces_full:
    dialogue=datset['test'][index]['dialogue']
    summary=datset['test'][index]['summary']
    prompt+=f'''
Dialogue:\n
{dialogue}
\nWhat was going on?
{summary}'''

  dialogue=datset['test'][example_index_to_summarize]['dialogue']
  prompt+=f'''
Dialogue:\n
{dialogue}
\nWhat was going on?'''
  return prompt
```

```python
example_indeces_full=[40]
example_index_to_summarize=200
one_shot_prompt=make_prompt(example_indeces_full,example_index_to_summarize)
print(one_shot_prompt)
```

```python
summary=datset['test'][example_index_to_summarize]['summary']
inputs=tokenizer(one_shot_prompt,return_tensors='pt')
output=tokenizer.decode(model.generate(inputs['input_ids'],max_new_tokens=50,)[0],skip_special_tokens=True)
print(dash_line)
print(f"Baseline Human Summary:\n{summary}\n")
print(dash_line)
print(f"Model Generation-One shot:\n{output}")
```

Few Shot Inference

```python
example_indices_full =[20,40,120]
example_index_to_summarize=200
few_shot_prompt=make_prompt(example_indices_full,example_index_to_summarize)
print(few_shot_prompt)
```

```python
summary=datset['test'][example_index_to_summarize]['summary']
inputs=tokenizer(few_shot_prompt,return_tensors='pt')
output=tokenizer.decode(model.generate(inputs['input_ids'],max_new_tokens=50,)[0],skip_special_tokens=True)
print(dash_line)
print(f"Baseline Human Summary:\n{summary}\n")
print(dash_line)
print(f"Model Generation-One shot:\n{output}")
```
